**Exploratory Data Analysis**

  Exploratory Data Analysis (EDA) is the process of exploring, investigating and gathering insights from data using statistical measures and visualizations. The objective of EDA is to develop and understanding of data, by uncovering trends, relationships and patterns.

EDA is both a science and an art. On the one hand it requires the knowledge of statistics, visualization techniques and data analysis tools like Numpy, Pandas, Seaborn etc. On the other hand, it requires asking interesting questions to guide the investigation and interpreting numbers & figures to generate useful insights.



    

  **Data Preprocessing**


   Data preprocessing refers to the set of techniques used to clean, transform, and prepare raw data for analysis or modeling. We need to clean the data , Handling the missing values and also removing the irrelevant attributes.It is a crucial step in the data science process to ensure that the data is in a suitable format for analysis or machine learning models.

```python
data=data.drop_duplicates()
data.duplicated().value_counts()
```

```python
data.drop('Genres',axis='columns',inplace=True)
```

```python
data['Type'].unique()
```

```python
data['Category'].nunique()
```

```python
data['Category'].unique()
```

```python
data['Reviews'] = pd.to_numeric(data['Reviews'],errors='coerce')
data['Rating'] = pd.to_numeric(data['Rating'],errors='coerce')
data['Price'] = pd.to_numeric(data['Price'],errors='coerce')
```

```python
data.info()
```

```python
data['Rating']=data['Rating'].fillna(0)
data['Rating']
```

```python
sorted((data['Rating'])[-5 :])
```

```python
data['Current Ver']=data['Current Ver'].fillna(0)
data['Android Ver']=data['Android Ver'].fillna(0)
data['Reviews']=data['Reviews'].fillna(0)
data['Type']=data['Type'].fillna(0)
data['Content Rating']=data['Content Rating'].fillna(0)
data['Price']=data['Price'].fillna(0)
```

```python
data.info()
```

**To check number of apps in each category**

```python
category=data['Category'].value_counts()
category
```

```python
plt.figure(figsize=(18,8))
plt.xticks(rotation=70)
plt.title('Number of Apps in each category')
plt.xlabel('category')
plt.ylabel('Total no.of apps in each category')
sns.barplot(x = category.index, y = category,hue=category)
```

**Distribution of app ratings**

```python

px.histogram(data, x ='Rating',nbins=20,title='Distribution of App Ratings')
```

```python
plt.figure(figsize=(18,8))
plt.hist(data['Rating'],bins=20)
plt.xlabel('Ratings',fontsize=20)
plt.ylabel('Count',fontsize=20)
```

**Investigating best or worse category based on ratings**

```python
category_ratings=data.groupby('Category')['Rating'].mean().sort_values(ascending=False)
category_ratings
```
